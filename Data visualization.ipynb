```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
am = pd.read_csv('D:\Data Visualization\Automobile.csv')
print(am.head())
```

```python
sns.distplot(am['city_mpg'], kde=False, rug=True)
plt.show()
```

```python
sns.jointplot(am, x='engine_size', y='horsepower')
plt.show()
```

```python
sns.jointplot(am, x='engine_size', y='horsepower', kind="hex")
plt.show()
```

```python
sns.jointplot(am, x='engine_size', y='horsepower', kind="reg")
plt.show()
```

```python
sns.jointplot(am, x='engine_size', y='horsepower', kind="resid")
plt.show()
```

```python
sns.jointplot(am, x='engine_size', y='horsepower', kind="kde")
```

```python
sns.pairplot(am[['normalized_losses', 'engine_size', 'horsepower']])
```

```python
sns.stripplot(am, x='fuel_type', y='horsepower')
```

```python
sns.swarmplot(am, x='fuel_type', y='horsepower')
```

```python
sns.boxplot(am, x='number_of_doors', y='horsepower', hue=am['fuel_type'])
```

```python
sns.barplot(am, x='body_style', y='horsepower')
```

```python
sns.barplot(am, x='body_style', y='horsepower', hue=am['engine_location'])
```

```python
sns.countplot(am, x='body_style', hue=am['engine_location'])
```
